# Por que usar Scikit-Learn?


> proporciona un gran numero de algoritmos de aprendizaje automatico. Esto es importante ya que programar algoritmos de aprendizaje automatico desde cero no es una tarea facil. La mayoria de los modelos en scikit-learn tambien tienen valores por defecto razonables para los hiperparametros. esto significa que un algoritmo de aprendizaje automatico puede que funcione bien con un poco de ajuste

> tiene un API limpio, uniforme y simplificado.

> tambien funciona bien con otras bibliotecas de python como numpy, pandas o matplotlib.

> en cuanto hayan entendido el uso basico y la sintaxis de scikit-learn para un modelo, cambiar a otro algoritmo sera muy sencillo.

> siempre habra mas tutoriales de scikit-learn. esto significa que es mas facil obtener las preguntas respondidas en sitios como stackoverflow

## es importante tener presente que ninguna herramienta es perfecta y que scikit-learn tiene algunas desventajas.

> 1. los modelos no se paralelizan tan bien como otras bibliotecas/marcos/lenguajes en los sistemas de produccion de aprendizaje profundo.

> 2. no se utiliza ampliamente scikit-learn para el aprendizaje profundo
