# Aprendizaje Supervisado

la forma mas comun de aprendizaje automatico.

Un algoritmo de aprendizaje supervisado aprende la relacion entre la matriz de caracteristicas y el vector objetivo para hacer predicciones.
imagen.png

una caracteristica es solo una propiedad de los datos que se representa como una columna. Un objetivo es la columna del conjunto de datos para la que se quieren hacer predicciones.

En cuanto un modelo aprende la relacion entre una matriz de caracteristicas ( se le llama matriz a multiples columnas) y un vector objetivo (se le llama vector a una sola columna), se puede hacer predicciones para datos ocultos o futuros.

En general, se puede pensar que el aprendizaje supervisado resuelve dos tipos de tareas diferentes.


## Regresion
> Predecir un valor continuo. Esto se considera un problema de regresion. Esto significa que el vector objetivo contiene cantidades continuas como los precios de las viviendas

imagen.png

## Clasificacion
>  Predecir un valor categorico. Esto se considera como un problema de clasificacion. Significa que el vector objetivo contiene cantidades categoricas como las diferentes especies de flores.

# Aprendizaje no supervisado

En el aprendizaje automatico no siempre se trata de predecir un valor.

Aveces el objetivo es encontrar alguna estructura en el conjunto de datos.

El aprendizaje **no supervisado ocurre cuando se preparan algoritmos sin darle respuestas para los ejemplos en el conjunto de datos**.

En el contexto scikit-learn, esto significa que solo se proporciona una **matriz de caracteristicas** cuando se ajusta el algoritmo.

**una matriz de caracteristicas es una cuadricula de dos dimensiones de datos donde las filas representan las muestras y las columnas representan las caracteristicas.**

al contrario del aprendizaje supervisado, no hay un vector objetivo.

imagen.png

Los algoritmos no supervisados no hacen predicciones a partir de los datos. Existen dos tipos comunes de algoritmos de aprendizaje no supervisado.

>1. Agrupamiento. suele usarse para descubir agrupaciones naturales en un conjunto de datos. 

Un uso popular de agrupamiento es en la segmentacion de mercado. Las empresas suelen tener grandes cantidades de informacion de sus clientes. al agrupar los clientes en diferentes segmentos, pueden vender o comercializar mas eficazmente a sus clientes.
imagen.png

>2. Reduccion de dimensionalidad. Son algoritmos de compresion de datos. Pueden hacer que los datos ocupen menos espacio en la computadora. 

Tener menos caracteristicas en los datos tambien puede hacer que sea mas facil la visualizacion de datos, asi como tambien acelerar el ajuste de los algoritmos de aprendizaje automatico
