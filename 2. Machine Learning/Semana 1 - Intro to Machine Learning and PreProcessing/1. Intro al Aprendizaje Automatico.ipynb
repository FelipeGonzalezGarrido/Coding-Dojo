## a continuacion se muestra un ejemplo del por que el Aprendizaje automatico es util

Digamos que tienen 150 flores. 

Estan interesados en determinar (clasificar) la especie de cada flor basandose en una caracteristica como la longitud de su petalo. 

Hay tres diferencies especies de flor en este conjunto de datos(tres clases): 

Iris setosa, 

Iris versicolor

Iris virginica.

Al analizar los datos, se puede escribir manualmente las reglas para separar las flores entre si. Esto no es aprendizaje automatico, ya que ustedes estan escribiendo las reglas.


### Regla 1.

## La longitud del petalos es menor o igual a 2,45?

De alli pueden hacer otra regla para tratar de separar versicolor de viginica

## Regla 2.


### La longitud del petalo es menor o igual a 4,45?


imagen.png

Asi, se pueden seguir escribiendo las reglas hasta que se hayan separado completamente las especies de flores entre si.

## Pero algo mas complicado, con mas caracteristicas, el programa deberia tener una lista muy larga con reglas complejas.

esta forma de escribir las reglas manualmente puede parecer factible para algunas aplicaciones como la clasificacion de flores.

### Pero que pasa cuando los datos cambian mas rapido de lo que se pueden actualizar las reglas?
### Puede este enfonque funcionar con problemas como reconocimiento de imagen o de voz ?

resulta que estos son los tipos de problemas donde los humanos no tienen necesariamente una buena comprension de como resolver el problema con enfoques basados en reglas.

aqui es donde brillan los enfoques de aprendizaje automatico.
