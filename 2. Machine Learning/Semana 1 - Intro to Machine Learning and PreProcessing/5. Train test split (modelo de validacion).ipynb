El objetivo del aprendizaje supervisado es crear un modelo que funcione bien con  **nuevos** datos. es asi como se puede observar como funciona en ellos.

El problema es que puede que no se tengan datos nuevos, sin embargo, se puede **simular** esta experiencia con un procedimiento de **validacion**.

El modelo de validacion se refiere al proceso donde un modelo entrenado se evalua con un conjunto de datos de prueba.

El modelo de validacion utilizado en esta seccion se llama **train test split**.

## 1. Divide el conjunto de datos en dos partes:

conjunto de entrenamiento (train): consta de aproximadamente un 75% de los datos.


conjunto de prueba (test): consta de aprox. un 25% de los datos

## 2. Solo entrenen el modelo del conjunto de entrenamiento.

## 3. Prueben el modelo en el conjunto de prueba y evaluen el rendimiento para simular como lo harian en los nuevos datos en los que NO se ha entrenado

# Fuga de datos

puesto que estamos simulando datos futuros con nuestro conjunto de prueba, es importante que no utilicemos ninguna informacion a partir de los **datos de prueba** en ningun momento hasta que estemos listos para probar nuestro modelo.

si utilizaramos informacion a partir de los datos de prueba o del conjunto de datos en general que incluyen datos de prueba, provocariamos una fuga de datos.

**La fuga de datos invalida la evaluacion final de nuestro modelo.**

Para este conjunto de datos, MedHouseVal sera el objetivo que estamos tratando de predecir. Por ahora, mantengamos todo el resto de las columnas como positbles caracteristicas.

```python
#El objetivo que intentamos predecir.
y = df["MedHouseVal"]

#Las caracteristicas que se usaran para hacer la prediccion.
X = df.drop(columns="MedHouseVal")
```

En cuanto los hayan cargado en el DataFrame y dividido en objetivo (y), y caracteristicas (X), pueden dividirlos en un conjunto de entrenamiento y conjunto de prueba

```python
#Importar train test split de sklearn
from sklearn.model_selection import train_test_split
# train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

random_state = 42, puesto que la division (datos de entrenamiento vs datos de test) es un proceso aleatorio, pordemos asegurarnos de obtener resultados reproducibles asignando un estado aleatorio.

si es 42, cada ves que se indique el mismo valor dara los mismos resultados

```python
#Para verificar la division, se comprueba con la funcion len()
len(X_train)
# arroja con el 75% de los datos
```
