La suposicion de todos los modelos predictivos es que hay una relacion entre las caracteristicas (X) y el objetivo (y). Esto se llama la senial. Un modelo trata de encontrar esa senial.

Sin embargo, tambien hay ruido en los datos. Esto es una aleatoriedad causada por factores distintos a las caracteristicas de nuestros datos. Este ruido se considera irreducible, ya que no es parte de la senal verdadera. Quizas si nuestras caracteristicas incluyeran toda la informacion del universo, podriamos eliminar el ruido. Sin embargo, en la realidad, un conjunto de datos siempre tendra ruido irreducible.

La imagen de abajo muestra la senal en la relacion entre x e y, Un modelo con un sesgo bajo y uan varianza baja puede encontrar esa senal o algo cerca de ella. Sin embarg, esa senal no se alinea perfectamente con todos los datos. Pero si reduce el error. El error restante es irreducible y se llama ruido.

# Identificar el subajuste

Sabran que el modelo esta subajustado cuando realiza datos de entrenamientoy de prueba deficientemente.

Por ejemplo, cuando evaluaron la puntuacion de R^2 del modelo en los conjuntos de entrenamiento y de prueba, encontraron que ambos estaban alreadedor de 0,35. O quizas el error absoluto medio es mas grande que el valor medio de la columna objetiva.

# Causas del subajuste

1. Un modelo demasiado simple
2. No hay suficientes datos
3. Las caracteristicas no correlacionan bien con el objetivo

Por ejemplo, si crean un modelo de arbol de decision con una profundidad maxima de 2, (en un modelo simple), para predecir la temeperatura de manana usando los precios de las acciones de hoy, (las caracteristicas no correlacionan con el objetivo), y un conjunto de entrenamiento de 5 muestras, (no son datos suficientes), es probable que el modelo tenga un sesgo alto.



# Reducir los sesgos:

1. Aumentar la complejidad del modelo
2. Anadir mas datos
3. Anadir caracteristicas con una alta correlacion al objetivo

Un modelo con una varianza alta puede hacer buenas predicciones usando algunos datos, pero no con otros. 

Un modelo con uan varianza alta tiene un sobreajuste en los datos de entrenamiento. 

Se ajusta al ruido que no es parte de la senal verdadera. 

Un modelo sobreajustado no sera capaz de ahcer buenas predicciones en los nuevos datos, como un conjunto de prueba. 

Es preciso con algunos y no con otros. 

Una varianza alta suele significar que pueden mejorar el rendimiento de sus modelos en el conjunto de prueba, que es lo mas importante.

Sabran que sus modelos estan sobreajustados en los datos de entrenamiento y en la varianza alta si el modelo funciona mucho mejor con los datos de entrenamiento que con los de prueba.

# Identificar el sobreajuste

Quizas la puntuacion R2 en los datos de entrenamiento es 0,90 y la puntuacion R2 en los datos de prueba es solo 0,5. o tal vez el RECM en los datos de entrenamiento es mucho mas bajo que el RECM en los de prueba.



# Causas del sobreajuste:

1. Demasiada complejidad en un modelo
2. No hay suficientes datos.
3. los datos de entrenamiento no son unas muestras representativas de la poblacion, todos los datos nuevos que el modelo pueda encontrar.

# Combatir el sobreajuste

1. Disminuir la complejidad del modelo
2. Agregar regularizacion
3. Agregar mas datos al conjunto de entrenamiento.
4. Garantizar qeu el conjunto de entrenamiento es un subconjunto representativo de todos los datos que el modelo encontrara

#El sesgo y la varianza son malos

Si trabajan como ingenieros de aprendizaje automatico, se esforzaran por equilibrar y disminuir tanto el sesgo como la varianza. Un modelo ideal hace menores errores tanto con los datos de prueba como los de entrenamiento. **Un sesgo alto significa que el modelo no ha encontrado la senal**, y una **varianza alta significa que puede haber un modelo que puede funcionar mejor en los datos de prueba.**
