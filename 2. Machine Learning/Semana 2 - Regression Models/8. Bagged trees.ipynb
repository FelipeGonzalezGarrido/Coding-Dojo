```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
# Import the bagging regressor
from sklearn.ensemble import BaggingRegressor
```

```python
filename = "/content/drive/MyDrive/Data CodingDojo/cali_housing.csv"
df = pd.read_csv(filename)
df.head()
```

```python
df.info()
```

```python
X = df.drop(columns="MedHouseVal")
y = df["MedHouseVal"]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

#1. Importar el modelo que quieran usar, BagginRegressor

```python
# from sklearn.ensemble import BagginRegressor
```

#2. Hacer una instancia del modelo
Este es un lugar donde podemos afinar los hiperparametros de un modelo, pero por ahora, vamos a utilizar los parametros por defecto.

```python
bagreg = BaggingRegressor(random_state = 42)
```

Hiperparametros

```python
bagreg.get_params()
```

Se puede observar que los parametros por defecto como n_estimators = 10. Este es solo uno de los parametros importantes que podrian explorar

# 3. Entrenar el modelo en los datos, almacenando la informacion aprendida de los datos.

```python
bagreg.fit(X_train, y_train)
```

#4. Predecir los valroes para y (recuerden que este paso permite ver las predicciones reales, pero no es necesario evaluar o afinar el modelo)

```python
bagreg.predict(X_test)
```

#5. Evaluar rendimiento del modelo

```python
bagreg_train_score = bagreg.score(X_train, y_train)
bagreg_test_score = bagreg.score(X_test, y_test)
print(bagreg_train_score)
print(bagreg_test_score)
```
