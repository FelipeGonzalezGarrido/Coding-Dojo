# Que son los bagged tress ?

bagging es la combinacion de dos palabras : bootstrap y aggregating

**bootstrappnig es el proceso de muestreo aleatorio con el uso del reemplazo**

Para ilustrar el concepto, finjan que los datos originales son los digitos 1-8. El diagrama de abajo muestra como estos dat os podrian ser meustreados por bootstrap en 5 muestras, cada una de ellas compuesta por 4 puntos de datos. Debido que el muestreo ocurre CON el reemplazo, los mismos puntos de datos (en este caso, el mismo digito) pueden aparecer muchas veces en una sola muestra.

imagen.png

Un modelo separado se entrena para cada una de las muestra. Cada muestra es un conjunto de entrenamiento separado. Cda modelo sera ligeramente diferente y puede hacer diferentes predicciones basandose en las muestras con las que se ha entrenado.
### Aggregating es el proceso de combinar resutlados individuales en un conjunto

El proceso final es agregar o combinar los resultados para cada modelo en una prediccion final. En el caso de un problema de regresion, la prediccion promedio de cada modelo se usa para una prediccion final.

imagen.png

#Ventajas
1. Funcionamiento tipicamente mejor
2. Estimaciones mas fiables de la importancia de las caracteristicas que los arboles de decision(en las futuras clases se hablara de la importancia de las caracteristicas)
3. Se puede usar para la clasificacion o regresion
4. No requiere escalamiento de caracteristicas como otros algoritmos que aprenderemos (KNN, regresion logistica, etc)
5. Se puede usar para problemas de clasficacion multiclase(mas de 2 clases)

#Desventajas
1. Son menos interpretables que los arboles de decision
2. Son mas lentos de entrenar
3. Son mas lentos para predecir
4. Los arboles individuales se pueden correlacionar entre si.
