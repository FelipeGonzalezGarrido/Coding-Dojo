#Modelos de ensamble

La idea detras de los modelos de ensamble es combinar los resultados de diversos modelos para superar los defectos de un solo modelo.

Algunos tipos de tecnicas de modelizacion de ensamble son:

* Bagging
* Boosting
* Stacking

Ya hemos hablado sobre bagging cuando discutimos sobre los bagged trees en la semana anterior. Ahora hablaremos sobre boosting, que es una gran herramienta para usar cuando necesiten aumenter el rendimiento del modelo, ya que tiende a disminuir el sesgo.

#**Modelos boosting**

La idea principal de los modelos boosting es utilizar secuencialmente los resultados de los aprendices debiles para finalmente mejorar el rendimiento del modelo final. Se utiliza para reducir el sesgo de modelizacion.

#**Potenciacion del gradiente**

Es una tecnica de boosting popular que ha ganado muchas competiciones de Kaggle, dado que tiene el potencial de producir excelentes resultados en datos tabulares.

Funciona secuencialmente ajustando modelos en los errores de los modelos anteriores.
