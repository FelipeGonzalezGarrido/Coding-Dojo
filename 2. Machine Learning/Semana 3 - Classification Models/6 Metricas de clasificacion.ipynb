Los modelos de clasificacion y regresion tratan diversas preguntas, y por lo tanto se deben evaluar de manera diferente. A continuacion se presenta un breve resumen de la diferencia entre la regresion y la clasificacion

#**Metricas de regresion vs de clasificacion**

**Metricas de regresion**

En un modelo de regresion una etiqueta objetivo podria tener cualquier valor, teoricamente.

Cuando estamos creando un modelo de regresion, tratamos de crear un modelo que predice una etiqueta que se acerque lo mas posible al valor de la etiqueta real de una muestra. Esta es la razon por la que utilizamos las metricas como un error abosulo medio, un error cuadratico medio o una raiz del error cuadratico medio. Queremos saber que tan lejos esta la prediccion del valor actual. De hecho, puede que nuestro modelo nunca haga una prediccion perfectamente exacta, y eso esta bien siempre y cuando este bastante cerca.


**Metricas de clasificacion**

Con los modelos de clasificacion, cada muestra es un miembro de uno de los numeros finitos de clases. Para cada muestra, el modelo predice la clase correcta o predice una de las clases incorrectas. Es correcto o incorrecto; no existe algo "cercano"

Debido a esto necesitamos diversas metricas. En esta actividad exploraremos como evaluar un modelo de clasificacion usando:

1. Exactitud
2. Precision
3. Memoria
4. Matriz de confusion
