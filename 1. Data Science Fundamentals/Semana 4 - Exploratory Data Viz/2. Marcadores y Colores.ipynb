```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
filename = "/content/drive/MyDrive/Data CodingDojo/athleteEventsNoPersonal.csv"

df = pd.read_csv(filename)

df.head()
```

# Numero de atletas por anio

Antes de avanzar con las visualizaciones, es necesario organizar el dato para Matplotlib.

esto significa transformar los datos en una lista python o un array NumPy.

el numero de valores unicos en una columna se puede encontrar con la funcion .nunique().

en este caso nos permitira saber la cantidad de ID por anio en el evento

```python
#Preparamos la variable con un groupby del dataframe, por anio mostrando los ID unicos
numberUniqueYear = df.groupby("Year")["ID"].nunique()
```

```python
#lo anteriro genera una Serie Panda
numberUniqueYear
```

# Establecer el tipo de marcador y color

empecemos creando un grafico basico donde uniqueYears esta en el eje X y numOlympian esta en el eje Y

```python
# Con la serie creada anteriormente (anio, cantidad de atletas) se crea el eje X para graficar
uniqueYears = list(numberUniqueYear.index)
uniqueYears
```

```python
# siendo el eje X el index (anio) de la serie creada, el eje Y entonces correspondera al numero de atletas 
numOlympian = list(numberUniqueYear.values)
numOlympian
```

# Tipo de marcador

```python
plt.style.use('seaborn')
plt.plot(uniqueYears, numOlympian)
```

image.png

```python
plt.plot(uniqueYears, numOlympian, marker = '.', markersize = 10)
```

# Cambio de color

image.png

```python
plt.plot(uniqueYears, numOlympian,c = 'magenta', marker = '.', markersize = 10)
```

```python
# Using hex strings
# '#000000' is black
# '#0000FF' would be blue
plt.plot(uniqueYears, numOlympian,c = '#000000', marker = '.', markersize = 10)
```

```python
# Using rgb tuples
# (0, 0, 0) is black
# (0, 0, 1) is blue
plt.plot(uniqueYears, numOlympian,c = (0, 0, 0), marker = '.', markersize = 10)
```
