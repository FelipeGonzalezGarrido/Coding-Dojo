```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
filename = "/content/drive/MyDrive/Data CodingDojo/athleteEventsNoPersonal.csv"

df = pd.read_csv(filename)

df.head()
```

Manipulacion de datos para su visualizacion

```python
# get number of olympians per year
numberUniqueYear = df.groupby(['Year'])['ID'].nunique()
```

```python
numberUniqueYear
```

```python
# Can convert to NumPy array or list
uniqueYears = np.array(numberUniqueYear.index)
uniqueYears
```

```python
numOlympian = np.array(numberUniqueYear.values)
numOlympian
```

```python
plt.style.use("seaborn")
```

```python
# Plot graph showing how it would look with defaults
plt.plot(uniqueYears, numOlympian, c= 'k')
```

# Estilo Matlab

```python
# Set xlim and ylim
plt.plot(uniqueYears, numOlympian, c= 'k')
plt.xlim(left=1892,right=2020)
plt.ylim(bottom=0,top=1800)
```

```python
# Set xlabel and ylabel
plt.plot(uniqueYears, numOlympian, c= 'k')
plt.xlim(left=1892,right=2020)
plt.ylim(bottom=0,top=1750)
plt.xlabel('Year')
plt.ylabel('Number of Olympians')
```

```python
# Set title
plt.plot(uniqueYears, numOlympian, c= 'k')
plt.xlim(left=1892,right=2020)
plt.ylim(bottom=0,top=1750)
plt.xlabel('Year')
plt.ylabel('Number of Olympians')
plt.title('Number of Olympians Over the Years')
```

```python
# Change fontsize
plt.plot(uniqueYears, numOlympian, c= 'k')
plt.xlim(left=1892,right=2020)
plt.ylim(bottom=0,top=1750)
plt.ylabel('Number of Olympians')
plt.title('Number of Olympians Over the Years')
plt.xlabel('Year', fontsize = 16)
plt.ylabel('Number of Olympians', fontsize = 16)
plt.title('Number of Olympians Over the Years', fontsize = 16)

```
