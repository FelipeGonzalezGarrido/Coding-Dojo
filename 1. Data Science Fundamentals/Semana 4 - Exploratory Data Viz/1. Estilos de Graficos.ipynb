# El estilo por defecto generalmente no suele ser el mas agradable

se puede utlilizar plt.style.available para mostrar diferentes esteticas a aplicar

```python
import matplotlib.pyplot as plt
```

```python
plt.style.available
```

# Estilo por defecto

```python
plt.style.use('classic')
plt.plot([10, 20, 30, 40, 50], [5, 10, 15, 20, 25])
```

```python
plt.style.use('fivethirtyeight')
plt.plot([10, 20, 30, 40, 50], [5, 10, 15, 20, 25])
```

```python
plt.style.use('ggplot')
plt.plot([10, 20, 30, 40, 50], [5, 10, 15, 20, 25])
```

```python
plt.style.use('tableau-colorblind10')
plt.plot([10, 20, 30, 40, 50], [5, 10, 15, 20, 25])
```

```python
plt.style.use('seaborn')
plt.plot([10, 20, 30, 40, 50], [5, 10, 15, 20, 25])
```
