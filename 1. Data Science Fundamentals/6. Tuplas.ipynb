# Tuplas

#### Son una secuencia ordenada de elementos, similares a las listas. con la diferencia que las tuplas no se pueden modificar. son Inmutables.

## Para inicializar una tupla vacia
Se puede hacer de dos formas.

```python
emptyTuple = ()

emptyTuple = tuple()
print(emptyTuple)
```

#### Para inicializar una tupla con valores tambien existen dos formas

```python
z = (3,7,4,2) # con los valores dentro de parentesis.
print(z)
```

```python
z = 3,7,4,2 # con los valores sin parentesis, ordenados y separados por coma.
print(z)
print(type(z))
```

```python
#Para crear una tupla de un solo valor, se debe agregar una coma al final, despues del elemento.
tup1 = ("Felipe",)
tup2 = ("Felipe")
print(type(tup1),type(tup2))
```

```python
# Para acceder al valor de una tupla se realiza igual que una lista. con indices

z = (3,7,4,2)
print(z[0])
# con la indexacion negativa se puede acceder al elemento final con -1
print(z[-1])
```

## Porciones de Tupla - SLICING

sliceTuple.png

```python
z = (3,7,4,2)
print(z[0:2]) # el primer indice es inclusivo pero el segundo indicado no lo es.
```

```python
print(z[:3]) # todo hasta el indice 3 pero sin incluir el indice 3
```

```python
print(z[-4:-1]) # tambien se puede hacer SLICING con indices negativos
```

## Las tuplas son inmutables, sus elementos no pueden ser modificados pero si pueden combinarse 2 tuplas o mas para obtener una tercera

```python
tup1 = ("Python","SQL")
tup2 = ("R",)
new_tuple = tup1 + tup2
print(new_tuple)
```

### Empaquetado y desempaquetado de tuplas

```python
# Todos los valores de una tupla se consideran empaquetados
z = (3,7,4,2)
```

```python
# Unpack each value in tuple
# first item in tuple will be assigned to a
# second item in tuple will be assigned to b
# third item in tuple will be assigned to c
# fourth item in tuple will be assigned to d
a, b, c, d = (3, 7, 4, 2)
print(a)
print(b)
print(c)
print(d)
```

## Enumerar

```python
friends = ('Steve', 'Rachel', 'Michael', 'Monica')
for index, friend in enumerate(friends):
    print(index,friend)

```

### Metodos de tuplas

para comenzar se inicializara la siguiente tupla

```python
animals = ('lama', 'sheep', 'lama', 48)
```

#### Index
nos indica el primer indice en el que aparece un valor.

```python
print(animals.index("lama"))
```

#### Count
Devuelve el numero de veces que un valor aparece en una tupla

```python
print(animals.count("lama"))
```
