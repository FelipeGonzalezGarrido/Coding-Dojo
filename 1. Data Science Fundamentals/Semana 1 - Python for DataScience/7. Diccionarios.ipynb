# Diccionarios
Son estructuras de datos que asignan claves unicas a valores {key:value}
#### **las claves de diccionario puede ser cualquier tipo de dato inmutable**
**numeros, cadenas,tuplas, etc**

si se asigna como valor un tipo de dato mutable, aparecera error
dictionariesRedBlue.png

```python
# Dictionaries are written within curly brackets {}. 
webstersDict = {'person': 'a human being',
               'marathon': 'a running race that is about 26 miles',
               'resist': 'to remain strong against the force',
               'run': 'to move with haste; act quickly'}
```

## Para acceder a un valor de diccionarios se usan corchetes, (como si fuera una lista)

```python
webstersDict["marathon"] # se busca el valor del key "marathon"
```

## Para agregar o actualizar una clave
se indica el key ( nuevo o existente) con los corchetes y luego su valor.

si el key es nuevo se crea en el diccionario, si ya existe el key, su valor sera reemplazado

```python
webstersDict["shoe"] = "an external covering for the human foot" # nuevo  key: valor
webstersDict
```

```python
webstersDict["marathon"] = "26 mile race" # valor existente sera actualizado
webstersDict
```

### Eliminar una clave del diccionario

Es imposible eliminar una clave y su valor correspondiente de un diccionario usando "del"

```python
del webstersDict["resist"] # se elimina clave y valor indicado
webstersDict
```

## Metodo get
Devuelve un valor a una clave dada. si la clave no existe devuelve None

```python
# Define a dictionary
storyCount = {'is': 100,
              'the': 90,
              'Michael': 12,
              'runs': 5}
storyCount.get('Michael') # ya que el KEY "michael" existe, devolvera su valor. 12.
```

```python
# tambien es util para buscar claves que no se sabe si existen, evitando obtener un error
# si no existe devuelve None y no un error
print(storyCount.get('chicken'))
```

```python
# se puede elegir el valor a devolver en caso de que la clave no exista.
print(storyCount.get("chicken", "no existe"))
```

### Metodo items()
devuelve el objeto de vista del diccionario. Se puede convertir la vista de diccionario como objeto en una lista utilizando la funcion lista  (list)

esto hara una lista con tuplas (key,value).

util para el desempaquetado (como tupla) de un diccionario

```python
list(webstersDict.items() )
```

```python
a,b,c,d,e,f = list(webstersDict.items() )
print(a[0],"-",a[1])
print(b)

```

```python
students = [
         {'first_name':  'Michael', 'last_name' : 'Jordan'},
         {'first_name' : 'John', 'last_name' : 'Rosales'},
         {'first_name' : 'Mark', 'last_name' : 'Guillen'},
         {'first_name' : 'KB', 'last_name' : 'Tonel'}
]

def iterateDictionary(students):
  for x in range(len(students)):
    a,b = students[x].items()
    print(a[0]+" - "+ a[1]+", "+ b[0] +" - "+ b[1])

iterateDictionary(students)
```
