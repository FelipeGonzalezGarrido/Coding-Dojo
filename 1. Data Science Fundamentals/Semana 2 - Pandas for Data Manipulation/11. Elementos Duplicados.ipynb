# Pandas Duplicated

Often tiems you wil have a series with duplicate values and you will want to know where they are.


```python
my_series = pd.Series(["Kanye West","Drake","Mac Miller","Drake","Beyonce","Kanye West","Drake"], name="artist")
my_series
```

Looks like Kanye West and Drake have duplicate values in the series above. This is an easy example. but what if you have 100k data points?

you will need a quicker way to locate duplicates than eye-balling it

# There are 3 ways to find duplicates
the first

the last

all of them

```python
# The First
series_duplicates_first = my_series.duplicated(keep="first")
series_duplicates_first.name = "duplicates"
series_duplicates_first
```

to get a better view of which one is duplicated, instead of True or False

```python
pd.concat([my_series, series_duplicates_first],axis=1)
```

```python
# The Last
series_duplicates_last = my_series.duplicated(keep="last")
series_duplicates_last.name = "duplicates"
series_duplicates_last
```

```python
pd.concat([my_series, series_duplicates_last],axis=1)
```

```python
# All Duplicates
series_duplicates_false = my_series.duplicated(keep=False)
series_duplicates_false.name = "duplicates"
series_duplicates_false
```

```python
pd.concat([my_series, series_duplicates_false],axis=1)
```
