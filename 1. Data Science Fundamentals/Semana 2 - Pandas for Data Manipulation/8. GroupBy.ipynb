```python
import pandas as pd
```

```python
filename = "/content/drive/MyDrive/Data CodingDojo/passenger_modified.csv"

df = pd.read_csv(filename)

df.head(2)
```

```python
columnas_nuevas = df.columns.str.lower().str.replace(" ", "_")
df.columns = columnas_nuevas
df.head()
```

```python
df[df["departure_delay_in_minutes"] == 1592]
```

```python
# evaluacion del promedio en comodidad del asiento
df['seat_comfort'].mean()
```

## Para obtener la evaluacion promedio de la comodidad del asiento, por clase.
## se puede utilizar GROUP BY

```python
df.groupby("class")["seat_comfort"].mean()
```

### Group BY tambien es utilizable en multiples columnas

```python
df.groupby(["class","gender"])["seat_comfort"].mean()
```
