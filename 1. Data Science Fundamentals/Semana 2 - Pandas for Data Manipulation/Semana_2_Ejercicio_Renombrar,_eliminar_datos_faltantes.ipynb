```python
from google.colab import drive
drive.mount("/content/drive")
import pandas as pd

filename = "/content/drive/MyDrive/Data CodingDojo/RailsToTrails_National Count Data_week.xlsx"
```

```python
bike_df = pd.read_excel(filename)
bike_df.head()
```

# 2. ¿Cuántos valores faltantes hay en el conjunto de datos? ¿En qué columnas están? 

```python
#VALORES FALTANTES
#En columnas: (2021 counts, Unnamed: 6, Unnamed: 7, Unnamed: 8, Unnamed: 9)
bike_df.isna().sum()
```

#1.Renombren la columna Unnamed: 6 a percent_change. 


```python
#CAMBIO DE NOMBRE DE COLUMNA Unnamed: 6
bike_df = bike_df.rename(columns={"Unnamed: 6":"percent_change"})
bike_df.head()
```

#3. Rellenen los valores faltantes en la columna 2021 Counts con 0. 

```python
#RELLENO DE VALORES COLUMNA 2021 Counts
bike_df["2021 Counts"].fillna(0, inplace=True)
bike_df["2021 Counts"].head()
```

#Eliminen las columnas innecesarias: 

```python
#Eliminacion de columnas innecesarias
bike_df = bike_df.drop(columns=["Unnamed: 7","Unnamed: 8", "Unnamed: 9"])
bike_df.info()
```
