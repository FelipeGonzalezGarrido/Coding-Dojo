```python
filename = "/content/drive/MyDrive/Data CodingDojo/mortgages.csv"
df = pd.read_csv(filename)
```

### Existen dos formas de renombrar columnas

## 1. Sustitucion de diccionario. 
###util para pocas columnas

## 2. Reemplazo de lista.
### requiere una lista completa de nombres.

```python
df.head()
```

```python
# cambio de nombre mediante sustitucion de diccionario. con funcion .rename
df = df.rename(columns={"Starting Balance": "starting_balance",
                        "Interest Paid": "interest_paid",
                        "Principal Paid": "principal_paid"})
df.head()
```

```python
# cambio de nombre mediante reemplazo de lista con metodo .columns
# es necesario conocer previamente las columnas para proceder con este metodo
df.columns = ["month",
              "starting_balance",
              "repayment",
              "interested_paid",
              "principal_paid",
              "new_balance",
              "mortgage_name",
              "interest_rate"]

df.head()
```

## Eliminacion de columnas

### 1. Metodo drop
permite eliminar multiples columnas a la vez

### 2. del 

```python
# eliminacion mediante drop
df = df.drop(columns=["new_balance"])

df.head()
```

```python
# eliminacion mediante -del-
del df["starting_balance"]
df.head()
```
