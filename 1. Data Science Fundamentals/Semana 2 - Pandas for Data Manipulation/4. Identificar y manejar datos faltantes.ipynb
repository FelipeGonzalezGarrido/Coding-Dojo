# Eliminar Filas
## Una regla de oro es que si menos de un 2% de las filas le faltan datos, seria mejor eliminarlas.

```python
df.dropna(inplace=True) # la funcion dropna(inplace=True) eliminara todas las filas a las que le falte algun dato. no se recomienda si existen muchos null
# nombre, edad, comentario
# felipe, 31  , Nan          esta columna sera eliminada al tener comentario NUll
# chimuelo, 4  , perro
```

```python
df.dropna(subset=["Embarked"], inplace=True)# con subset=["columnaX"] se especifica donde estan los null a eliminar para actuar sobre ellos y no en todo el dataset
df.isna().sum()
```

## Ventajas de eliminar filas

1. No se introduciran ningun error a los datos. Cada vez que se rellenan datos faltantes con valores que no estan seguros de que sean correctos, se esta introduciendo errores en los datos.

2. las filas restantes guardan toda su informacion. si eliminan columnas cada fila pierde alguna informacion que podria ser util.

## Desventajas de eliminar filas
1. Se perderan datos inevitablemente al eliminar filas.

## Eliminar columnas

1. si en la exploracion de datos se encuentran columnas con una gran cantidad de datos erroneos o faltantes, eliminar fila causara una perdida de informacion muy grande. para estos casos puede resultar mejor rellenar los datos, estando seguros de su procedencia o simplemente eliminar la columna.
2. la regla de oro con las columnas es. 
si faltan mas del 50% de los datos, puede ser eliminara. (Siempre previa evaluacion)

```python
#FUNCION .DROP(columns="x")
df.drop(columns="Cabin", inplace=True)

df.isna().sum()
```

# IMPUTAR VALORES FALTANTES
La imputacion significa adivinar que valores deberian estar en las celdas faltantes. esta siempre introduce errores a los datos y se deberia usar con precaucion y consideracion.

existen muchas formas de hacerlo, como tambien hay estrategias faciles con las que se puede empezar.


## La mejor forma de imputar valores faltantes es:

1. examinar con cuidado los datos para buscar patrones en los datos faltantes.
Hay correlaciones entre los valores faltantes en una columna con valores en la otra columna?

2. consultar a un experto en la materia o usar conocimientos propios en el tema para entender los datos y aprovechar ese conocimiento para hacer conjeturas sobre como rellenar los datos.

3. buscar por patrones en los datos que sugieren que los datos pueden pertenecer en las celdas faltantes.


# Rellenar los datos numericos faltantes con un promedio.

Una estrategia para rellenar datos numericos es hacerlo con un promedio de esa columna. No sera un valor exacto pero se acercara mas al valor correcto dada la informacion disponible.

para esto se utiliza la media, mediana o moda.

La media es una buena opcion para valores flotantes y la mediana o moda para valores enteros.


Rellenaremos la columna "Age" (edad) de nuestro conjunto de datos con una mediana para mantener todos los valores como numeros enteros.

```python
median_age = df["Age"].median() # almacenaremos en variable median_age el resultado de la funcion .median() de la columna "Age"
df["Age"].fillna(median_age, inplace=True)# ahora se llenanr los valores null con funcion fillna y la variable creada con funcion .median()
df.isna().sum()
```

# Rellenar datos categoricos con el valor mas comun de la columna

Cuando hacerlo?

Es buena estrategia si un valor es sumamente mas comun que los otros valores.
Ej: si una columna llamada "tipo de mascota" fuera 99% perros y 1% pescados, y no hay mayor pista sobre que serian los valores faltantes.
se puede asumir que los valores faltantes tambien pueden ser perros. de esta forma solo se estaria rellenando, en el peor de los casos, un 1% de informacion erronea.


el valor mas comun es sinonimo con moda y panda permite usar este metodo con datos categoricos.

a continuacion, se rellena los valores faltantes en la columna "Pet Type" con el valor mas comun en esa columna

```python
most_common_pet = df["Pet Type"].mode() # con funcion .mode() se obtiene el valor mas comun de la columna indicada
df["Pet Type"].fillna(most_common_pet, inplace=True) # se aplica el valor mas comun a los valores faltantes de la columna Pet Type
```

# Ventajas de imputar valores

1. no se pierde ningun dato.


# Desventajas de imputar valores

1. Se introducen errores en los datos.
