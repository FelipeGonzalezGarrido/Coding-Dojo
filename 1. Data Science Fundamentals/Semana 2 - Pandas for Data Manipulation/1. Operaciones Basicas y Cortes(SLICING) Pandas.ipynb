### Cual es la diferencia entre un array de numpy y un series de Pandas

Mientras que, hasta ahora, Series de Pandas siempre han empezado en 0, pueden tener índices que pueden ser cadenas e incluso si tienen valores números, no tienen que empezar en 0.
