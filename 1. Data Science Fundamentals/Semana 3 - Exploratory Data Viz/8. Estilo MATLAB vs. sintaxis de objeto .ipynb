```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
filename =  "/content/drive/MyDrive/Data CodingDojo/athleteEventsNoPersonal.csv"


df = pd.read_csv(filename)
df.head()
```

Pongan los datos en una forma adecuada para el grafico

```python
numberUniqueYear = df.groupby("Year")["ID"].nunique()
```

```python
numberUniqueYear
```

```python
uniqueYears = np.array(numberUniqueYear.index)
uniqueYears
```

```python
numOlympian = np.array(numberUniqueYear.values)
numOlympian
```

# Estilo MATLAB vs. sintaxis de objeto

## Estilo MATLAB

Se trata de una interfaz con scripts diseñada para que se sienta como si fuera MATLAB. Matplotlib mantiene un indicador a la figura actual y le envía comandos.

## Orientado a objetos

Esto se suele más usar en situaciones donde se quiere más control sobre la figura.

Nota importante: ustedes pueden y tendrán a menudo trazados que se crearán a través de una combinación de estilo MATLAB y la sintaxis orientada a objetos.

# Estilo MATLAB

```python
plt.style.use('seaborn')
plt.plot(uniqueYears, numOlympian, c= 'k')

```

#Orientado a objetos



```python
fig, axes = plt.subplots(nrows = 1, ncols = 1)
axes.plot(uniqueYears, numOlympian, c= 'k');

```

# Combinación

A menudo verán ambos tipos de sintaxis usados juntos. En el caso del código de abajo, los ejes son el código orientado a objetos y los comandos plt.xticks y plt.yticks son estilo MATLAB.

```python
fig, axes = plt.subplots(nrows = 1, ncols = 1);
axes.plot(uniqueYears, numOlympian, c= 'k');
plt.xticks(fontsize = 25);
plt.yticks(fontsize = 25);

```
