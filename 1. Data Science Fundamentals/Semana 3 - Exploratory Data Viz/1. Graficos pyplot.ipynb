```python
import pandas as pd
import matplotlib.pyplot as plt
```

## La funcion plt.plot() acepta una serie de estructuras de datos de python, en particulas las listas, las tuplas, los arrays numpy, y las columnas de dataframes de pandas.
es importante determinar que datos se quieren en cada grafico para luego asignarlos usando plt.plot(x,y)

```python
plt.plot([10, 20, 30, 40, 50], [5, 10, 15, 20, 25]) # 2 listas como eje x e y
```

```python
# You can also plot another line on the same graph
plt.plot([10, 20, 30, 40, 50], [5, 10, 15, 20, 25])
plt.plot([10, 20, 30, 40, 50], [100, 400, 900, 1600, 2500])

# matplotlib por defecto incluira en el mismo grafico las lineas indicadas en una misma celda de notebook
```
