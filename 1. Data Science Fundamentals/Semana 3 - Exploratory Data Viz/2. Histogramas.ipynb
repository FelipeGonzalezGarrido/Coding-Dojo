Es una practica comun crear histogramas para explorar los datos y obtener una idea general de como lucen.

### un histograma es un resumen de la variacion en una variable medida. Muestra el numero de ejemplos que ocurren en una categoria.

Un histograma es un tipo de distribucion de frecuencia.

funcionan dividiendo todo el rango de valores en una serie de intervalos y luego contando cuantos valores caen en cada intervalo

Un aspecto a mejorar del grafico mostrado es la notacion cientifica del eje X.

```python
# Try to remove scientific notation
df['price'].hist()
plt.ticklabel_format(useOffset=False, style='plain')
```

habiendo eliminado la notacion cientifica el valor de X se sobrelapa lo cual se corrige al rotar las etiquetas

```python
df["price"].hist()
plt.xticks(rotation = 45)
plt.ticklabel_format(useOffset=False, style="plain")
```

a continuacion, repartiremos los datos utilizando mas intervalos

```python
# Change number of bins
df['price'].hist(bins = 30)
plt.xticks(rotation = 45)
plt.ticklabel_format(useOffset=False, style='plain')
```

Ahora nos enfocaremos en un subconjunto d datos que solo recoge los precios de 300000 o menos, un contorno negro en cada intervalo para una mejor visualizacion

```python
price_filter = df.loc[:, 'price'] <= 3000000
df.loc[price_filter, 'price'].hist(bins = 30,edgecolor='black')
plt.xticks(rotation = 35)
plt.ticklabel_format(useOffset=False, style='plain')
```

Tambien se puede utilizar otro estilo, como el usado a continuacion llamado seaborn

```python
plt.style.use('seaborn')
price_filter = df.loc[:, 'price'] <= 3000000
df.loc[price_filter, 'price'].hist(bins = 30,edgecolor='black')
plt.xticks(rotation = 45)
plt.ticklabel_format(useOffset=False, style='plain')
```
