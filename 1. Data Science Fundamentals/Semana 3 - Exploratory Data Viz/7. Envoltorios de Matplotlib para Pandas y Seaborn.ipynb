```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
filename = "/content/drive/MyDrive/Data CodingDojo/athleteEventsNoPersonal.csv"

df = pd.read_csv(filename)


df.head()
```

```python
yearFilter = df["Year"] == 2016
top20Height = df.loc[yearFilter]
```

```python
yearFilter = df['Year'] == 2016
top20Height = df.loc[yearFilter, :].groupby(['NOC'])['Height'].mean().sort_values(ascending = False).head(20)
top20Height
```

MATPLOTLIB tiene una gran cantidad de visualizaciones pero no son las mas esteticas y funcionales, lo que significa es que ustedes suelen necesitar muchos códigos para hacer que las visualizaciones sean visualmente atractivas y funcionales.

Lo que se puede hacer para evitar estos problemas es usar la funcionalidad de trazado en la biblioteca de Pandas, que puede funcionar como un envoltorio para Matplotlib y la biblioteca Seaborn de Python, el cuál es mejor para trazados estadísticos más complicados

```python
# Convert data from pandas to plot using matplotlib
# Notice it isn't the most readable graph
plt.bar(top20Height.index, top20Height.values);
```

```python
# mas legible
plt.bar(top20Height.index, top20Height.values);
plt.xticks(rotation = 90); # permite manipular ubicacion de leyenda eje x
```

Uso de grafico de barras en pandas

```python
top20Height.plot.bar();
plt.ylabel('Average Height (cm)')
```

# Uso de sintaxis matplotlib

```python
top20Height.plot.bar();
plt.ylabel('Average Height (cm)')
```

En definitiva, los graficos de pandas son envoltorios de MATPLOTLIB que requieren menos codigo para realizarce, pero de requerir mas caracteristicas se debe recurrir a la sintaxis de matplotlib

como:

plt.ylabel("Average Height")  muestra la leyenda para eje Y

plt.xticks(rotation = 90);   rota la etiqueta del eje X

#Uso de grafico de barra en SEABORN

La integración de estructuras de datos de Pandas

API orientado al conjunto de datos para examinar las relaciones entre múltiples variables.

Ayuda especializada para usar variables categóricas para mostrar observaciones o agregar estadísticas.

Control conciso sobre el estilo de las figuras de Matplotlib con varios temas incorporados.

Herramientas para elegir gamas de colores que fielmente revelan patrones en los datos.

```python
# With seaborn, there is no need to convert data from pandas if you have a dataframe,
# however our data is a pandas Series, you need to convert it.
# Notice it isn't the most readable graph
sns.barplot(x = top20Height.index, y = top20Height.values);
plt.xticks(rotation = 90);
```

Seaborn es una gran libreria, aunque su mejores atributos resaltan cuando se trabaja con graficos estadisticos como heatmaps o boxplots
