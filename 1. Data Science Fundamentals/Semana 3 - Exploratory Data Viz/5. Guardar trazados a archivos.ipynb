```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
filename = "/content/drive/MyDrive/Data CodingDojo/linearRegPredicted.csv"
```

```python
df = pd.read_csv(filename)
df.head()
```

```python
feature = df["feature"].values
actual = df["actual"].values
predicted = df["predicted"].values
```

## Guardar trazados a archivos
Es importante guardar la visualizacion ya que siempre existe la posibilidad de que el grafico no se vea de la misma manera en el cuaderno en el archivo de imagen.

# Estilo MATLAB

```python
# The legend will be cutoff when saved
plt.plot(feature, predicted, c = 'r', label = 'Prediction')
plt.scatter(feature, actual, c= 'k', label = 'Actual')
plt.legend(loc=(1.02,0))
plt.savefig('MATLABlegendcutoff.png', dpi = 300)
```

```python
# The legend will not be cutoff when saved
plt.plot(feature, predicted, c = 'r', label = 'Prediction')
plt.scatter(feature, actual, c= 'k', label = 'Actual')
plt.legend(loc=(1.02,0))
plt.tight_layout()
plt.savefig('MATLABSavedPlot.png', dpi = 300)

```

# Orientado al objeto
