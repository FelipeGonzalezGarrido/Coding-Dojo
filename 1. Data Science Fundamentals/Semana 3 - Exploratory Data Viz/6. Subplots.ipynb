```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

 En este caso, existe un modelo de regresión lineal que se ajustó con un intercepto y otro modelo que se ajustó sin un intercepto. 
 
 Pueden encontrar qué filas son parte de qué modelos de la columna de intercepción. También hay una columna de características. Pueden verlo como los datos que se usaron para hacer una predicción. 
 
 La columna actual es la que debería haber predicho un modelo de regresión lineal. La columna de predicción es lo que el modelo predijo.

Para que el aprendizaje de los subplots sea más interesante, el modelo con la intercepción se graficará en el subplot izquierdo y el modelo sin una intercepción se graficará a la derecha. 

Cada subplot tendrá la línea de mejor ajuste en rojo (característica y columna predicha) y el gráfico de dispersión de puntos negros utilizando la información de la característica y la columna real. Cada subplot contendrá un título que mostrará qué gráfico es cuál.

```python
filename = "/content/drive/MyDrive/Data CodingDojo/linearWithWithout.csv"

df = pd.read_csv(filename)

df.head()
```

```python
interceptFilter = df["intercept"] == True
df_intercept = df.loc[interceptFilter,:]
df_nointercept = df.loc[~interceptFilter,:]
```

# Subplots

Sirve para comparar subconjuntos de datos lado a lado.

```python
#ajustar el tamanio de la figura
plt.figure(figsize =(8,4))

#subplot 1

plt.subplot(1,2,1);
plt.plot(df_intercept["feature"].values, df_intercept["predicted"].values, c="r");
plt.scatter(df_intercept["feature"].values, df_intercept["actual"].values, c="k");
plt.title("intercept", fontsize = 12);
#subplot 2
plt.subplot(1,2,2);
plt.plot(df_nointercept["feature"].values, df_nointercept["predicted"].values, c="r");
plt.scatter(df_nointercept["feature"].values, df_nointercept["actual"].values, c="k");
plt.title("no intercept", fontsize= 12);
```
