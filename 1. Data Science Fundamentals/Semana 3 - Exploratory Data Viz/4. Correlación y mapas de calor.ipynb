# Correlacion

Como cientificos de datos, solemos estar interesados si las diversas caracteristicas de un conjunto de datos pudieran estar relacionadas.

Para datos numericos, una clave para describir una relacion es la correlacion entre los valores.

# si dos variables cambian en la misma direccion, tendran una correlacion positiva.

por ejemplo, hay una correlacion positiva entre las horas trabajadas y el sueldo.

a medida que aumentan las horas, aumenta el sueldo.

a medida que disminuyen las horas, disminuye el sueldo.


# Si dos variables cambian en la direccion opuesta, tendran una correlacion negativa.

por ejemplo, hay una correlacion negativa entre la altitud y la temperatura.

a medida que aumenta la altitud, disminuye la temperatura.

a medida que disminuye la altitud, aumenta la temperatura.

# No correlacion

no existe ninguna relacion entre la altura de una persona y el costo del pan.

# Grado de correlacion


tambien se puede cuantificar la relacion con un valor numero llamado "coeficiente de correlacion"

este valor mide la fuerza de la relacion y puede ir de -1 a 1.

si dos variables no se relacionan, tendran un coeficiente de correlacion de 0.

un valor 1 es la relacion positiva mas fuerte, un coeficiente de correlacion de 0.75 muestra una relacion mas fuerte que una de 0.45

un valor de -1 es la relacion negativa mas fuerte. un coeficiente de correlacion de -0.9 muestra una relacion mas fuerte que una de -0.7

```python
corr = df.corr()
```

una vez que estas correlaciones hayan sido calculadas, una gran manera de visualizarlos es a traves del uso de un mapa de calor.

se pueden elegir los colores que quieran usando (cmap="Blues" o cmap="Oranges")

```python
import seaborn as sns
sns.heatmap(corr, cmap="Greens")
```

tambien puede ser de utilidad agregar los coeficientes de correlacion actuales al mapa de calor usando annot=True

```python
sns.heatmap(corr, cmap="Greens", annot=True)
```

Observen que cada columna numérica se compara con la otra.  Cuando una columna se compara consigo misma, siempre es una perfecta correlación, es por eso que vemos los valores de 1 para ID con ID, edad con edad, altura con altura y año con año. 

En general, un coeficiente de correlación más grande que 0,7 se considera “fuerte”.

Entre 0,5 y 0,7 se considera “moderado”.  

Entre 0,3 y 0,5 se considera una correlación “baja”.

Sin embargo, estas reglas de oro pueden variar entre las áreas de contenido.

En la muestra de mapa de calor de arriba, debido a que todos los valores bajo 0,3 en magnitud, no pudimos identificar ninguna relación entre las variables en nuestro conjunto de datos.
