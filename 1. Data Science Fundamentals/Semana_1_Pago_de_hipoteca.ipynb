1) De cunto ser el pago mensual de una hipoteca a 30 aos (30 * 12 = 360 meses de plazo) con un capital de 400 000 dlares y un 3 % TAE (tasa de inters con comisiones incluidas)? 

```python
P = 400000
R = 0.03 /12
N = 30*12
resultado = P*(R*(1+R)**N)/((1+R)**N-1)
print(f"El pago mensual sera de: ${round(resultado,2)}")
```

2) Cunto es el pago mensual si la tasa de inters es del 6 %? (utilicen la misma duracin del plazo y el mismo principio)

```python
P = 400000
R = 0.06 / 12
N = 30*12
resultado = P*(R*(1+R)**N)/((1+R)**N-1)
print(f"El pago mensual sera de: ${round(resultado,2)}")
```
