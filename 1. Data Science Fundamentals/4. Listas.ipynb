Metodo count()

cuenta las ocurrencia del objeto indicado

```python
random_list = [4, 1, 5, 4, 10, 4]
random_list.count(4)
```

Metodo sort()

ordena de menor a mayor la lista indicada

```python
z = [3, 7, 4, 2]
z.sort()

z
```

Metodo append()

agregara al final de la lista el elemento indicado

```python
z = [7, 4, 3, 2]
z.append(3)
print(z)
```

Metodo remove()

eliminara la primera ocurrencia del valor indicado

```python
z = [7, 4, 3, 2, 3]
z.remove(2)

z
```

Metodo pop()

elimina un elemento en el indice que se proporciona.

adicionalmente devuelve el elemento eliminado de la lista.

por defecto elimina el ultimo valor de la lista

```python
z = [7, 4, 3, 3]
print(z.pop(1))
print(z)

```

Metodo extend()

amplia una lista agregando mas elementos, el beneficio de esto es que se pueden combinar listas

extendMethod.png

```python
z = [7, 3, 3]
z.extend([4,5])
print(z)
```

Alternativamente esto se puede lograr con el operador +

```python
print([1,2] + [3,4])
```

Metodo insert()

Inserta un elemento antes del indice que se proporcione

```python
z = [7, 3, 3, 4, 5]
z.insert(4, [1, 2])
print(z)
```
