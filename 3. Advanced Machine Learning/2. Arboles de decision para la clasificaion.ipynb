Empecemos nuestra exploracion de modelos de clasificacion con un modelo que conocemos de nuestro estudio de modelos de regresion: el arbol de decision.

El clasificador del arbol de decision funciona de la misma manera que el regresor del arbol de decision, salvo que en lugar de que una prediccion final sea un promedio de todos los valores en el nodo hoja, la prediccion final es la clase. Por ejemplo, un conjunto de datos muy comun que involucra la clasificacion de flores en una de las tres especies que se basan en las caracteristicas de las flores se muestra aqui. esta es una clasificacion multiclase donde las tres especies de flores son setosa, versicolor y virginica.
