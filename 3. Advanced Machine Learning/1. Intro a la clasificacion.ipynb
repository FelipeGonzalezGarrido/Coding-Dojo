# Clasificacion binaria o multiclase

La tarea mas simple de clasificacion es una clasificacion binaria.
es donde el modelo esta haciendo una prediccion entre dos clases.
por ejemplo, si o no. O si es perro o gato.

No todas las tareas de clasificacion son binarias, si hay mas de dos posibles clases lo denominamos una clasificacion multiclase.
ejemplo. predecir que tipo de comida cuando las opciones son leche, carne, vegetal o fruta.

# Clases balanceadas o desequilibradas

Otra caracteristica importante de un problema de clasificacion es evaluar que tan balanceadas estan las clases.
por ejemplo, en un problema de clasificacion binaria, un conjunto de datos perfectamente balanceado tendria la mitad de una clase y la mitad de la otra.

A medida que una clase se vuelve mas comun en el conjunto de datos , este se vueulve mas desequilibrado. si el 75% de los animales son gatos y el 25 perro. este conjunto esta desequilibrado.


# Modelos de clasificacion

algunos de los modelos que usamos para la regresion tiene una contraparte que se usa para la clasificacion. Por ejemplo. todos nuestros modelos basados en arboles tienen un regresor y un clasificador.

Tambien vamos a aprender un nuevo tipo de modelo llamado KNN, que se puede usar para la clasificacion y tambien para la regresion.

Y aprenderemos el modelo de clasificacion de nombre confuso: la regresion logistica.
Que aunque tenga la palabra regresion en el nombre, este modelo se usa para la clasificacion y NO para la regresion.
