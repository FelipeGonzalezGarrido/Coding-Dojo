# Tipos de error

En la clasificacion, existen dos formas basicas de equivocarse en la prediccion. Un error es un falso positivo, y otro es un falso negativo. Segun el conjunto de datos, un tipo de error puede ser mas problematico que otro.

Cuando utilicen los terminos "positivo" y "negativo", recuerden que esto no esta describiendo un resultado bueno o malo. Por ejemplo, todos hemos escuchado de resultados positivos por COVID. positivo significa que una persona tiene la enfermedad, lo que no es algo bueno...




Con respecto al COVID, dos errores pueden ocurrir cuando se realiza un prueba.

**observen que ambos terminos incluyen la palabra "falso" en la descripcion**

Una prueba de falso positivo significa que los resultados de la prueba de la persona muestran que tiene la enfermedad, pero en realidad, la persona no la tiene.

Una prueba de falso negativo significa que la prueba de la persona dice que NO tiene COVID, y que en realidad, la persona si lo tiene

Si bien estamos hablando de posibles resultados, tambien hay dos maneras de describir un resultado correcto en este caso:

**Observen que ambos terminos incluyen la palabra "verdadero" en la descripcion.**

un verdadero positivo significa que la prueba de la persona dice que es positiva y que SI tiene covid.

un verdadero negativo significa que la prueba de la persona dice que es negativa y que NO tiene covid.

El error de tipo I, tambien conocido como falso positivo

es un tipo de error comun. por ejemplo, es posible que hayan escuchado la frase:" los trabajadores estan tratando de evitar falsos positivos inclusi si eso significa mas falsos negativos". Esto significa que no quieren contratar accidentalmente trabajadores que resultan no ser buenos (falso positivo) incluso si eso significa que rechazaran a personas mas cualificadas (falso negativo), ya que esto es aparentemente mejor que a veces contratar personas no cualificadas.

El error de tipo II, tambien conocido como un falso negativo, es otro tipo de error comun donde se predice que algo negativo, cuando realmente es positivo.
