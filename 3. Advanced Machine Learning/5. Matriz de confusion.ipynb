# Matriz de confusion

una matriz de confusion es una tabla que se suele usar para describir el rendimiento de un modelo de clasificacion (o "clasificado") en un conjunto de datos de prueba cuyos valores verdaderos se conocen. En terminos simples, muestra que' fue predicho y que deberia haberse predicho. Es una matriz que muestra los verdaderos negativos, falsos positivos, falsos negativos y verdaderos positivos.

Puede tomar un poco de tiempo para procesar los resutlados de una matriz de confusion, asi que no hay prisa.

Un mapa de calor muestra todos los resultados de la prediccion de un modelo se llama matriz de confusion. Crearan una matriz de confusion para un modelo en el siguiente ejercicio. por ahora, evaluemos esta muestra de matriz de confusion.

esto muestra un total de las 179 predicciones ( 87+16+!7+59 = 179).

La etiqueta "true label" (etiqueta verdadera) esta en el eje y, y la prediccion en el eje "x"

En este caso, una prediccion de 1 se consideraria positiva, sin embargo, una prediccion de 0 se consideraria negativa.

- la casilla superior izquierda tenia una prediccion de 0 y en realidad era 0. Esto corresponde a los verdaderos negativos. Tenemos 87 
- la casilla superir derecha tenia una prediccion de 1 y en realidad era 0. Esto corresponde a falsos positivos. tenemos 16
- la casilla inferior izquierda tenia una prediccion de 0 y en realidad era 1. Esto corresponde a falsos negativos. tenemos 17
- la casilla superior derecha tenia una prediccion de 1 y en realidad era 1. Esto corresponde a verdaderos positivos. tenemos 59.

Deben ser capaces de determinar la precision del modelo basandose solo en la matriz de confusion. la precision es simplemente el numero de predicciones correctas.

las predicciones correctas fueron los verdaderos negativos y los verdaderos positivos. 87+59 = 146 prediccion correcta

para encontrar la precision, solo dividan las predicciones correctas por el total de predicciones y multipliquenlo por 100 para obtener el porcentaje...

146/179 = 0,816 o 81,6% de precision
