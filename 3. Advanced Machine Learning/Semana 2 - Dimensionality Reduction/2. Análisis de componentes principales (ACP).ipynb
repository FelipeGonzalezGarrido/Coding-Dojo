#**ACP**

El análisis de componentes principales, abreviado como ACP, es una de las técnicas más comunes que se utilizan para la reducción de dimensionalidad. Es una técnica de aprendizaje no supervisado, lo que significa que no utilizaremos un vector objetivo/y para esta técnica. 

ACP **transforma nuestras características en un nuevo conjunto de variables llamado componentes principales, que son combinaciones lineales de todas nuestras características originales**. El algoritmo se basa en gran medida en temas de álgebra y la comprensión de las matemáticas. Estos están más allá del alcance de este curso, pero si estás interesado en aprender más sobre cómo el algoritmo funciona detrás de escena, revisa las secciones opcionales al final de este capítulo.

Los componentes principales (CP) están ordenados en función de la variabilidad que explican: el primer CP explica la mayor parte de la varianza de nuestros datos originales, mientras que el último CP explica la menor cantidad de variación de nuestros datos originales. Elegimos un número selecto de CP que explican acumulativamente la varianza en nuestros datos y, sin embargo, reducen la dimensionalidad de nuestro espacio de características.

##**Usos de ACP**

ACP es útil para dos tareas principales:

- Visualización: si tenemos datos de alta dimensión eso sería imposible de visualizar, sin embargo, si transformamos estos datos en un espacio de baja dimensión, los podemos visualizar como clústeres.
- La reducción de dimensionalidad para mejorar la velocidad del modelo: podemos utilizar ACP para primero reducir las dimensiones de nuestros datos y luego utilizar CP para modelar.

##**Supuestos de ACP**

Cuando realizamos ACP, asumimos:

- Que hay relaciones lineales entre nuestras características.
-  Que la importancia se determina a partir de la varianza, es decir, que una varianza más alta significa que es más importante.
- Que todas nuestras características originales tienen las mismas unidades (si no los tienen, necesitas escalar tus datos).
