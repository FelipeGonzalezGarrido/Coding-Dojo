# Por que necesitamos la reduccion de dimensionalidad ?

Al trabajar con datos grandes, solemos trabajar con datos que tienen muchas columnas ( o muchas dimensiones). El aprendizaje automatico es una gran herramienta para usar con datos de alta dimension, ya que pueden detectar patrones con espacios de alta dimension que serian imposibles de detectar para los humanos. 

La "Maldicion de la dimensionalidad" se refiere al fenomeno que surge cuando se trabaja con espacios de alta dimension que aumentan. A medida que el numero de caracteristicas aumenta, la cantidad de datos necesarios para generalizar con precision un modelo a nuevos datos crece exponencialmente.

La reducción de dimensionalidad se refiere a cualquier ténica donde tratemos de reducir la característica de espacio en la que estamos trabajando. En general, existen dos tipos de reducción de dimensionalidad:

- Selección de la característica

¡Ya has estado haciendo una selección de la característica! Este es el proceso de selección de un subconjunto de características para modelar. Por ejemplo, si estás creando un modelo para predecir el precio de casas, puedes elegir utilizar solo las características más importantes en tu modelo final y omitir las características que no sean importantes para el modelo.

¿Pero si no sabes qué características elegir?

- Extracción de la característica

La extracción de la característica refiere a las técnicas donde se toman todas las características y se combinan en ciertas maneras para reducirlas en dimensiones más pequeñas.

Algunos métodos populares de extracción de la característica incluyen:

  - Análisis de componentes principales (ACP)

  Algoritmo de marchine learning no supervisado


  - Análisis discriminante lineal (ADL)
