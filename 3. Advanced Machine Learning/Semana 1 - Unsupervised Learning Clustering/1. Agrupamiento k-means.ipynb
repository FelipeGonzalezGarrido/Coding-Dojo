#Aprendizaje no supervisado

Es usado cuando no se tiene un objetivo en el conjunto de datos y necesitamos que la maquina aprenda los patrones en los datos dados.

**Agrupamiento**

Es una tarea comun de aprendizaje no supervisado donde se intenta agrupar los datos en distintas agrupaciones o grupos basandose en caracteristicas o patrones comunes.

Por ejemplo, es posible que tengan un conjunto de datos de todos los clientes y quieran personalizar sus esfuerzos de marketing en funcion de sus segmentos de cliente. Puede agrupar los datos para determinar diferentes grupos de clientes.

Suelen haber patrones en nuestros datos que no podemos observar, sin embargo, los algoritmos de agrupacion si los pueden captar.

#**Agrupamiento k-means**

K-MEANS es un algoritmo de agrupamiento popular. Generalmente funciona asi:

- Escoger aleatoriamente los puntos K para comenzar como nuestros centros de agrupamiento (comunmente llamados centroides)
- Para cada punto de datos, asignar cada uno de estos a uan agrupacion basado en el centro al que esta mas cerca.
- Actualizar los centros de agrupamiento cuando se encuentra la media de todos los puntos de datos en cada agrupamiento.
- Repetir los pasos 2 y 3 hasta la convergencia (hasta que los centros dejen de cambiar y los puntos generalmente se queden en el mismo agrupamiento)
