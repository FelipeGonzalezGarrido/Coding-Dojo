#DBSCAN

DBSCAN (agrupamiento espacial basado en densidad de aplicaciones con ruido) es un algoritmo que se utiliza para el agrupamiento.

Este es un algoritmo útil para agrupar los clústeres de alta densidad, al tiempo que se ignoran los puntos de datos que pueden no encajar en un clúster (denominados "ruido"). Elegimos dos hiperparámetros dentro de este algoritmo:

- Épsilon o eps (ϵ): Puedes pensar de esto como un radio que el algoritmo buscar alrededor del punto de datos para considerar agregar el punto de datos a su clúster. Es la distancia mínima entre dos muestras para que una se considere como la vecina de la otra. Dos puntos se consideran ser el mismo clúster si la distancia entre los dos puntos está debajo del umbral de la épsilon.
- min_samples: el número mínimo de puntos que se necesitan para crear un clúster.


El algoritmo funciona al:

1. Escoger un punto de datos aleatorio para estar en un clúster.
2. Mirar los puntos dentro de la distancia eps a partir del punto en el paso 1. Si hay al menos un número de puntos de datos de min_samples, estos se asignan a ese clúster.
3. Luego, examina cada punto del clúster y los puntos que se encuentran a una distancia eps de cada punto se añaden al clúster.
4. Repetir el paso 3 hasta que no haya más vecinas entre la distancia eps de cada punto en el clúster.
5. Repetir los pasos 1-4 para cada clúster adicional.
6. Los puntos que no cumplen los criterios de distancia o min_samples se marcan como “ruido” y no se incluyen como parte de un clúster.

Una ventaja de DBSCAN es que no necesitas elegir el numero de clusteres antes de tiempo. Encontrara el numero optimo de clusteres por si mismo. DBSCAN tambien produce mejores resultados que otros algoritmos de agrupamiento como k-means o punto de datos donde la densidad es mas explicativa para los clusteres que la distancia.
