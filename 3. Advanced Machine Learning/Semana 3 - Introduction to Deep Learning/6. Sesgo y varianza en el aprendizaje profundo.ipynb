# Como podrás recordar:

La varianza en un modelo de rendimiento se relaciona con el sobreajuste.  Significa que un modelo funciona bien en los datos que entrenó, pero no puede generalizar ese aprendizaje a los nuevos datos:

El sesgo en el modelo de rendimiento se relaciona con el subajuste.  Significa que un modelo no se ha ajustado con éxito a los datos de entrenamiento y no puede realizar buenas predicciones en los datos de entrenamiento o de prueba.


#Complejidad del modelo
- Subajuste: no es lo suficientemente complejo

imagen.png

La imagen de arriba muestra el historial de entrenamiento de un modelo subajustado.  La exactitud del modelo en los conjuntos de entrenamiento y de prueba es baja y no está mejorando.

Si tu modelo está subajustado y no puede funcionar bien incluso en los datos de entrenamiento, puede significar que necesites más complejidad.

Puedes agregar más capas (generalmente es lo mejor)

O puedes agregar más neuronas (a veces es mejor)

Más complejidad suele causar un bajo sesgo, pero una varianza potencialmente mayor.

# Sobreajuste: demasiado complejo

imagen.png

La imagen de arriba muestra el historial de entrenamiento de un modelo sobreajustado.  Observa que la exactitud del modelo en los datos de entrenamiento continúa mejorando, pero la exactitud de los datos de prueba no.

Una fuente común de sobreajuste es el aprendizaje profundo es demasiada complejidad.  Esto significa un modelo con demasiadas ponderaciones totales.  

Más capas → más complejidad

Más neuronas → más complejidad

Si tu modelo está sobreajustado, una forma de combatir eso es hacerlo más pequeño.

Menos complejidad suele causar un alto sesgo, pero puede causar una baja varianza.

Otra herramienta importante para combatir la alta varianza en la regularización, como las capas de abandono.  Aprenderás sobre esto en una futura clase.

# Entrenamiento y sobreentrenamiento

Otra fuente de varianza es el sobreentrenamiento.  Esto significa que estás entrenando tu modelo con demasiadas o muy pocas épocas.  

imagen.png

En el gráfico de modelo de entrenamiento anterior, podemos ver que la exactitud del modelo en el conjunto de prueba comienza a caer en aproximadamente 10 épocas.  El modelo estaba sobreentrenado.  Una forma para combatir el sobreentrenamiento es utilizar una llamada de parada temprana, de la que aprenderás en una futura clase.
#Resumen:

Para reducir el sesgo:

- Aumenta el número de capas de tu modelo
- Aumenta el número de neuronas en cada capa
- Entrena tu modelo para más épocas 

Para reducir la varianza:

- Elimina las capas
- Disminuye las neuronas
- Entrena el modelo durante menos épocas
- Agrega regularización (como las capas de abandono)
