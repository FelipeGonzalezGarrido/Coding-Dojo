Vocabulario:

- Nodo: una unidad que procesa la información entrante y produce un solo número de salida 

- Capa: una colección de nodos que trabajan en paralelo 

- Capa de entrada: el primer conjunto de nodos que trabajan sobre las características de la muestra de entrada 

- Capa de salida: el último conjunto de nodos que emiten una predicción 

- Propagación hacia delante: el procesamiento y la transmisión de información hacia adelante a través de todas las capas para producir una predicción de salida 

- Función de costo: la diferencia entre las predicciones y todas las muestras y sus etiquetas verdaderas 

- Propagación hacia atrás: el proceso de actualizar las ponderaciones de cada nodo para reducir la función de costo

# Introducción

Los modelos secuenciales transmiten información de una capa de nodos a la siguiente un cierto número de veces hasta que producen una salida.  A estos también se les llaman redes prealimentadas.

Existen muchos tipos de redes secuenciales que traspasan información a través de diferentes tipos de capas.  Lo que los hace secuenciales es que la información se transmite y se transforma en una dirección desde la entrada hasta la salida.  Esto se llama propagación hacia delante.

La primera clase de modelo que aprenderemos se llama red densamente conectada, o red densa.  En una red densa cada nodo en una capa traspasa su saluda a todos los nodos en la capa siguiente, como se visualiza más abajo.

## Nodos

Una forma de pensar sobre cada nodo en una red densa es como un modelo de regresión lineal separado con una función de activación adicional aplicada a la salida.  De hecho, ya has visto algo parecido a esto, ¡una regresión logística!  Una regresión logística es un modelo de regresión lineal con una función sigmoide aplicada a la salida.

imagen.png

Un algoritmo de regresión lineal encuentra una línea de mejor ajuste que describe la función de una o más entradas X a la salida y del objetivo.  Hace esto multiplicando cada característica de salida x por un coeficiente aprendido.  En el aprendizaje profundo lo llamamos coeficientes de ponderación.  Los productos de las entradas y las ponderaciones se agregan junto con una intercepción, el cual se llama sesgo en el aprendizaje profundo.

##Funciones de activación

Cada nodo aplica una función de activación a la suma de los productos de las entradas y las ponderaciones y el término de sesgo.  Existe una variedad de funciones de activación que un ingeniero en aprendizaje automático puede intentar, y cada capa puede tener una función de activación diferente.  Algunos ejemplos son: lineal, sigmoide, tangente hiperbólica (tanh) y unidad lineal rectificada (ReLU).
##Capas

Una capa son múltiples nodos que aprenden cada uno sus propias ponderaciones y sesgos para aplicar los datos de entrada.  Los nodos en una capa funcionan en paralelo y no transmiten información a cada uno directamente.  Después de aplicar una función de activación elegida por el creador, cada nodo en una capa transmite su salida a cada nodo en la próxima capa.  Las salidas de la capa anterior se convierten en las entradas para la siguiente capa.

#Capa de entrada

Los nodos en la primera capa, llamados la capa de entrada, cada uno recibe una copia de las características de los datos de entrada.  Cada nodo tiene diferentes ponderaciones que se aplican a cada característica y a un sesgo diferente para agregar cuando se suman.  Sin embargo, cada nodo en una capa utiliza la misma función de activación.

#Capas ocultas

Entre la capa de entrada y la capa de salida hay un número cualquiera de capas ocultas, cada una con su propia colección de nodos.  Las entradas de cada nodo en esas capas son las salidas de todos los nodos en la capa anterior.  Si la capa de entrada tiene 10 nodos, cada nodo en la primera capa oculta tendrá 10 ponderaciones para aplicar a las salidas de cada uno de los nodos en la capa de entrada.

#Capa de salida

La capa de salida es el último conjunto de nodos que se aplicará.  Su salida de esta capa es la salida de la red entera.  Debido a esto, la capa de salida tiene un número especial de nodos y una función de activación especial según el tipo de problema de aprendizaje automático.  El número de nodos en la capa de salida determina cuantas salidas de la red produce, y la función de activación determina qué tipo de salida la red produce.  

Por ejemplo:

- Si solo un número continuo está en la salida adecuada, la capa de salida tendrá un nodo y utilizará una función de activación lineal.  
- Por otro lado, si se desearon dos salidas y deberían ser un 0 o un 1, la capa de salida tendrá dos nodos y la función de activación será sigmoide. 

Reglas fáciles para las funciones de activación de la capa de salida:

- Regresión: utiliza una función lineal
- Clasificación binaria: utiliza sigmoide
- Clasificación multiclase: utiliza softmax 

# Determinación de la función de costo:

En el proceso de entrenamiento, o ajuste, que es decir encontrar las ponderaciones correctas para cada nodo en cada capa, la red procurará predecir un objetivo muchas veces.  Cada vez que realiza un paso de propagación hacia delante y produce una salida, esa salida se comprará a una etiqueta verdadera para esa muestra.  La diferencia entre la red de salida y la etiqueta verdadera es el error, y los errores de todas las muestras se llama la función de costo.  
# Propagación hacia atrás

En cuanto las predicciones se hayan hecho en una o más muestras, los errores, llamados colectivamente la función de costo, se usan para actualizar las ponderaciones en cada capa para prepararlas para el siguiente paso de propagación hacia delante.  El proceso para actualizar las ponderaciones se llama el paso de propagación hacia atrás.  La propagación hacia atrás intenta reducir la función de costo para producir más predicciones exactas.
# Época

El proceso de completar un paso de propagación hacia delante en cada muestra en el conjunto de entrenamiento y de actualizar las ponderaciones de cada nodo con una propagación hacia atrás se llama época.
#Resumen:

La propagación hacia delante es el proceso de introducción de características de una muestra a una capa de entrada y la propagación de información hacia delante a través de algunos números de capas donde cada una procesa la información de alguna manera.  La propagación hacia delante termina con una salida a partir de una capa de salida.  Una red neuronal realiza un número definido de los pasos de propagación hacia adelante y hacia atrás para encontrar las mejores ponderaciones para cada nodo en la red para hacer predicciones exactas.
