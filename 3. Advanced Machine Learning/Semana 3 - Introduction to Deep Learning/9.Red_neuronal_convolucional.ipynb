##Introducción a CNN

Si ves una imagen de un perro, es muy probable que puedas determinar que es de hecho una imagen de un perro y no de un gato. Sin embargo, esta es una difícil tarea para una computadora, ya que estas no tienen la misma habilidad de visualizar imágenes como nosotros.

image.png

Las redes neuronal convolucionales (CNN) revolucionaro la vision por computadora debido a lo buenas que son al trabajar con los datos de imagen. Si bien se pueden utilizar para otros tipos de datos, es mas comun usarlas con datos de imagen debido a su eficiencia de trabajar con grandes cantidades de datos.

## Trabajo con datos de imagen

Al trabajar con imagenes, necesitamos una manera de convertir imagenes como las vemos en un formato que las computadoras puedan reconocer. Para hacer esto, cada pixel en la imagen se convierte en un valor en un array. Para las imagenes a color, cada pixel tiene un valor de color, lo que es cierta cantidad de rojo, verde y azul.

image.png

## CNN

Las CNN funcionan utilizando capas convolucionales y capas de agrupacion.

Las capas convolucionales se basan en la operacion matematica de [convolucion](https://en.wikipedia.org/wiki/Convolution), donde los filtros se utilizan en secciones de la imagen a la vez. El efecto que esto tiene es que cada capa de una red puede primordialmente detectar diferentes componentes de imagen. Tal vez una capa detecta bordes, una detecta ojos, otra detecta oidos, etc. Para finalmente clasificar la imagen.
