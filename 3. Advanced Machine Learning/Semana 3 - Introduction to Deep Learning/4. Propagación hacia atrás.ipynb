Has aprendido sobre la propagación hacia delante donde un modelo hace una predicción, y sobre la función de costo que indica al modelo lo lejos que están las predicciones de las etiquetas verdaderas.

Pero ¿cómo sabe el modelo ajustar las ponderaciones para hacer una mejor predicción?

Vocabulario:

- Nodo: una unidad que procesa la información entrante y produce un solo número de salida 

-  Capa: una colección de nodos que trabajan en paralelo 

-   Capa de entrada: el primer conjunto de nodos que trabajan sobre las características de la muestra de entrada 

-  Capa de salida: el último conjunto de nodos que emiten una predicción 

-  Propagación hacia delante: el procesamiento y la transmisión de información hacia adelante a través de todas las capas para producir una predicción de salida 

-  Yhat: las predicciones de un modelo para todas las muestras en X. 

-   Costo: la combinación de todos los errores de un modelo en todas las muestras. 

-   Función de costo o función de pérdida: la diferencia entre las predicciones y todas las muestras y sus etiquetas verdaderas 

-   Propagación hacia atrás: el proceso de actualizar las ponderaciones de cada nodo para reducir la función de costo 

-    Época: un círculo completo de la propagación hacia delante y propagación hacia atrás para todas las muestras de entrenamiento.

#Modelo como una función:

Un modelo predictivo es una función.  Una función es un amplio término que describe todo lo que toma una entrada y produce una salida única.  Numerosas entradas pueden producir la misma salida, pero cada entrada puede solo producir una salida.

Un modelo predictivo asigna una salida, X, con una predicción, yhat.  

  

Pero puesto que sabemos los valores reales de las etiquetas, y, podemos agregar la función de costo a la función de modelo predictivo para hacer una nueva función que asigna todas las muestras de las entradas de la muestra X a la función de costo.

> f(X) = y - g(X)

#Un pequeño cálculo:

Existe una herramienta muy genial en cálculo llamada ecuaciones diferenciales.  Una ecuación diferencial nos puede decir cómo la salida de una función cambia por una sola entrada determinada al encontrar el derivado de esa función.  Con esta herramienta podemos determinar cómo necesitamos cambiar el modelo para cambiar la función de costo.  Es decir, nos dice cómo las ponderaciones en la capa de salida necesitan cambiar para reducir el error total que el modelo está haciendo con todas las muestras.
#Descenso de gradiente

Cambiar las ponderaciones para reducir la función de costo se llama descenso de gradiente.  Se llama así porque podemos visualizar la función de costo como un mapa topográfico donde los índices de error más bajas están “abajo”.  Cambiar las ponderaciones para reducir la función de costo es como una pelota rodando cuesta abajo.

La imagen anterior muestra cómo, con cada época, el modelo reduce la función de costo, o los errores que hace en las predicciones.  No sabe cuál es REALMENTE la función de costo, las ecuaciones diferenciales solo le dicen, por un valor determinado de las ponderaciones, qué camino es hacia abajo y cómo cambiar la ponderación para reducir el costo.  No sabe qué valor para la ponderación producirá realmente el menor error.  Por lo tanto, lo intenta una y otra vez, cada vez cambiando la ponderación un poco para reducir el costo, como una pelota rodando cuesta abajo.

La imagen anterior muestra un descenso de gradiente para una sola ponderación, pero nuestros modelos tienes numerosas ponderaciones para numerosas entradas.  La función de costo gana una dimensión para cada ponderación en un nodo.

Si tuviéramos 2 diferentes ponderaciones para cambiar, la función de costo y el descenso de gradiente se verían más como la imagen de abajo.  El cálculo tiene acuaciones para encontrar cómo debe cambiar cada ponderación cuando hay numerosas entradas en una función.  Estas se llaman derivadas parciales, ya que solo son pare de la derivada de la función entera.  En backpropagation el modelo utiliza ecuaciones diferenciales parciales para entender cómo cambiar cada ponderación individualmente para alcanzar el descenso de gradiente.

#Propagación hacia atrás

Ya vimos cómo funciona el descenso de gradiente para ajustar las ponderaciones en la capa de salida, pero ¿qué hay de las capas antes de él?  El algoritmo para el backpropagation nos permite determinar cómo cada capa antes de la capa de salida necesita cambiar basándose en cómo la que está delante de ella cambió para reducir la función de costo.  Los cambios en las ponderaciones se propagan hacia atrás a través de capas.

Si deseas conocer el algoritmo completo, puedes revisar [esta explicación por Shane de Silva.](https://towardsdatascience.com/the-maths-behind-back-propagation-cf6714736abf)

# Resumen:

En cuento el paso de propagación hacia delante de un modelo ha hecho predicciones para todas las muestras, las predicciones se comparan con los valores reales para cada muestra para producir un costo.  El costo la combinación de todos los errores que el modelo hizo en todas las muestras.  La función que relaciona las entradas del modelo con el costo se denomina función de costo.  El modelo utiliza un descenso de gradiente para reducir el costo al ajustar las ponderaciones en cada capa comenzando con la capa de salida y se mueve hacia atrás, hacia la capa de entrada.  Un círculo completo de la propagación hacia delante y propagación hacia atrás se llama época.  Un modelo pasa por numerosas épocas, cambiando sus ponderaciones internas cada vez para reducir el costo.
