El aprendizaje profundo es un subconjunto de aprendizaje automático que se ha convertido muy popular en los últimos años debido a su rendimiento junto con la mejora de la velocidad, la memoria y la potencia de la computadora.

Existen numerosos tipos de redes neuronales artificiales, como las de avance, convolucionales y recurrentes. En general, las redes neuronales artificiales tienen una capa de entrada para admitir cada una de sus características, algunos números de capas ocultas y una capa de salida.

Se pueden utilizar para la regresion y clasificacion y trabajan muy bien con datos grandes.

Cómo se ajustan las redes neuronales

En un pase hacia delante de una red neuronal, ejecutamos nuestros datos a través de una serie de capas, cada de unas consistiendo en un número de neuronas o nodos. 

En general, un pase hacia delante a través de una red neuronal funciona por este proceso:

1. Toma nuestras entradas y multiplica casa entrada por una ponderación.
2. Suma los resultados del paso 1.
3. Suma un valor (llamado sesgo) al resultado del paso 2.
4. Pasa el resultado del paso 3 a través de una función de activación.
5. Luego se pasa el resultado del paso 4 a la capa siguiente, donde el proceso se repite. 

Las ponderaciones y sesgos se actualizan a través de un proceso llamado backpropagation. Puesto que esto se basa en gran medida en cálculos, no profundizaremos con muchos detalles, pero es importante para comprenderlo en un alto nivel. Revisa los recursos opcionales al final de este capítulo si desean profundizar más este tema.

Como cualquier algoritmo de aprendizaje automático, estamos tratando de minimizar la función de pérdida. En el caso de las redes neuronales, estamos tratando de encontrar los valores para las ponderaciones y sesgos que minimizan la función de pérdida. 

Una pasada completa hacia adelante y hacia atrás de todos nuestros puntos de datos se llama una época. Entre más épocas entrenamos en nuestro modelo, nuestro modelo hará un mejor trabajo al aprender las ponderaciones y los sesgos que minimizan la función de pérdida para nuestros datos de entrenamiento. 

Es importante tener en cuenta el sobreajuste, un problema común para las redes neuronales.
