#¿Por qué utilizar las funciones de activación?

Durante la propagación hacia delante, después de que las entradas de cada nodo se multiplican por las ponderaciones de nodos por estas entradas y luego se suman con el sesgo, el nodo aplica una función de activación antes de enviar su salida a la siguiente capa.

Las funciones de activación son importantes porque crean funciones no lineares para predecir clases o valores.  Sin una función de activación, un nodo es principalmente un algoritmo de regresión lineal.  Una regresión lineal solo puede aplicar una función lineal a un conjunto de entradas.  Si el problema no se puede modelar con una línea recta, una regresión lineal no puede modelarlo de manera eficaz.

Las funciones de activación hacen que las salidas de los nodos sean no lineares.  Esto les permite a los modelos de aprendizaje profundo encontrar soluciones no lineares en problemas complejos.

#Sigmoide:

La función sigmoide asigna todos los valores de salida del nodo a un valor entre 0 y 1, como se muestra arriba en el eje y del primer gráfico.  Esto es equivalente a las probabilidades previstas por un modelo de regresión logística con LogisticRegression().predict_proba_(X) en sklearn. 
#Tanh (o tangente hiperbólica)

La función tanh asigna la salida de un nodo a un valor entre -1 y 1, como se muestra arriba en el eje y del segundo gráfico.
#ReLU (unidad lineal rectificada)

La función de activación ReLU es muy popular entre los ingenieros de aprendizaje automático debido a que funciona bien.  Asigna todas las salidas negativas de un nodo a 0 y devuelve todas las salidas positivas tal cual (función lineal), como se muestra arriba en el tercer gráfico.  Se cree que el rendimiento excelente se debe a que algunos nodos se “apagan” al asignar sus valores a 0.  Esta es una forma de regularización similar a una expulsión, del cual aprenderán en otra clase.
#Funciones de activación en el desarrollo del modelo

Las funciones de activación son hiperparámetros establecidas por el ingeniero.  Parte del desarrollo del modelo es afinar las funciones de activación para la entrada y las capas ocultas para encontrar las que funcionan mejor.  ReLU es un buen lugar para comenzar.
#Funciones de activación de la capa de salida

La excepción es la capa de salida.  Las funciones de activación específicas se requieren para un modelo para producir clases de salidas específicas.

- Lineal (modelos de regresión): Si el modelo debe tener una salida continua que puede ser positiva o negativa.  Una función de activación lineal simplemente devuelve el valor que se le pasó. g(z) = z.

- Sigmoide (modelos de clasificación binaria): si el modelo solo saca un 0 o un 1, como en los problemas de clasificación binaria.

- Softmax (modelos de clasificación multiclase): si el modelo devuelve el candidato más probable a partir de una lista finita de opciones, como el problema de clasificación multiclase.

#Resumen

Las funciones de activación se aplican después que las ponderaciones y entradas se multiplican y suman junto al término del sesgo para cada nodo para determinar la salida del nodo.  Estos le permiten al modelo encontrar una función no lineal asignar las características de entrada a la etiqueta de objetivo.  Si bien las funciones de activación de la entrada y las capas ocultas son ajustables, la función de activación para la capa de salida depende del tipo de problema que es moldeado.
