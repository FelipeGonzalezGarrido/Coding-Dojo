**Introducción a RNN**

Las redes neuronales recurrentes (RNN) se utilizan para trabajar con datos secuenciales. Esto incluye

- Datos de series temporales
- Datos de lenguaje
- Datos de sonido
- Datos de video
- Y mucho más

La razón por la que utilizan las RNN en datos secuenciales es porque estas toman en cuenta lo que pasó previamente. Las RNN hacen esto al alimentar la salida del paso de tiempo anterior en la red en el siguiente paso de tiempo.

**LSTM y GRU**

Un problema con RNN es que, durante el proceso de ajuste, las redes completan lo que llamamos backpropagation con el tiempo. Esto implica multiplicar muchas gradientes (debido a la regla de cadena), lo que puede llevar al problema de la desaparición del gradiente. Las LSTM y GRU son ciertos tipos de capas recurrentes que revuelven estos problemas.
