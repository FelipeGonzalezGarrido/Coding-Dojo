Vamos a explorar otro modelo popular de aprendizaje automatico llamado KNN. Este modelo se puede usar para las tareas de clasificacion y regresion.

a medida que aprendan este modelo, observen que la prediccion se basa en la distancia de un punto de datos con otros puntos de datos similares. Por esta razon, tendremos que **escalar** nuestras caracteristicas para el uso con los modelos KNN.

# Que es KNN ?

Es un modelo que usa la "K" la mayoria de las observaciones similares para hacer predicciones. A continuacion se muestran algunas ventajas y desventajas:

##**Ventajas**
- Es facil de entender y explicar que otros algoritmos de aprendizaje automatico
- Se puede usar para la clasificacion o regresion
- Se puede usar para problemas de clasificacion de multiclase

##**Desventajas**
- Debe almacenar todos los datos en los que se ajusto
- Su fase de prediccion puede ser lenta cuando los datos sobre los que se ha "ajustado" son grandes.
- Suele tener un mal rendimiento que otros metodos de aprendizaje supervisado.



## Como funciona.

1. El usuario especifica el valor para K. En este ejemplo, elegimos k=5, vecinos alrededor del punto negro.
2. Busquen las observaciones k en los datos que mas se acercan a las mediciones de una muestra desconocida.
- La distancia euclidiana se suele usar como la distancia metrica.
3. Utilicen el valor objetivo mas popular a partir de k vecinos mas cercanos como el valor objetivo previsto. En el ejemplo, delos 5 vecinos mas cercanos del punto negro, 2 son cafes y 3 son verdes. Puesto que tenemos una mayoria de puntos verdes alrededor de este punto negro, le asignamos la etiqueta verde.

A excepcion del paso 3, el algoritmo funciona de forma similar a la regresion, salvo que la prediccion es el promedio del objetivo numerico del k vecinos mas cercanos. Observen que cando k es pequeno, KNN es una varianza alta con un algoritmo de sesgo bajo. Cuando k es grande, KNN es una varianza baja con un algoritmo de sesgo alto.



Para las imagenes a continuacion, cada region con color (naranja, cian y azul ) es una region de clasificacion. En una region, cada punto se clasificara como cierta clase. Imaginen que, si un punto esta en una region naranja, significa que se clasificara como una naranja real. Si un punto esta en la region cian o azul, se clasificara como otra cosa.

En este caso, los limites de la region son menos distintos para las tres clases cuando K es pequeno (varianza alto, sesgo bajo). Mientras que para una K grande, los limites de la region son mas distintos ( varianza baja, sesgo alto)
